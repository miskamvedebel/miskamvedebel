```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import pymc3 as pm
```

```python
sns.set(style='darkgrid', palette='muted')
```

```python
def simulate_linear_data(N, beta_0, beta_1, eps_sigma_sq):
    
    '''Simulating random dataset using noisy linear process'''
    
    arr = list(map(lambda x: float(x)/100.0, np.arange(100)))
    X = np.random.RandomState(42).choice(arr, N, replace=False)
    df = pd.DataFrame({'x': X})
    
    # y ~ beta_0 + beta_1 * x + epsilon
    # generating x
    
    eps_mean = 0.0
    df['y'] = beta_0 + beta_1 * df['x'] + np.random.RandomState(42).normal(eps_mean, eps_sigma_sq, N)
    
    return df
```

```python
beta_0 = 1.0
beta_1 = 2.0

N = 100
eps_sigma_sq = 0.5

df = simulate_linear_data(N=N, 
                          beta_0=beta_0, 
                          beta_1=beta_1, 
                          eps_sigma_sq=eps_sigma_sq)

sns.lmplot(x='x', y='y', data=df, size=10)
plt.xlim(0.0, 1.0)
```

```python
def glm_mcmc_inference(df, iterations=5000):
    
    '''Calculates the Markov Chain Monte Carlo trace of GLM Bayesian Linear Regression'''
    
    basic_model = pm.Model()
    with basic_model:
        
        pm.glm.GLM.from_formula("y ~ x", df, family=pm.glm.families.Normal())
        
        # finding MAP
        start = pm.find_MAP()
        step = pm.NUTS()
        trace = pm.sample(iterations, step, start, random_seed=42, progressbar=True)
        
        return trace
```

```python
trace = glm_mcmc_inference(df)
pm.traceplot(trace[:])
plt.show()
```

```python
#plotting sample of posterior regression lines

sns.lmplot(x='x', y='y', data=df, size=10, fit_reg=False)
plt.xlim(0.0, 1.0)
plt.ylim(0.0, 4.0)

#pymc3
pm.plot_posterior_predictive_glm(trace, samples=200)
#true line
plt.plot(np.linspace(0, 1, N),
        beta_0 + beta_1*np.linspace(0, 1, N), 
        lw=3.,
        c='green',
        label='True Regression Line')
plt.legend(loc=0)
plt.show()
```
