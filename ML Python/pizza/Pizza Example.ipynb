```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
import randomizer
data = pd.read_csv('pizza_dataset.csv', sep=';')
data_with_country = pd.read_csv('pizza_dataset_with_country.csv', sep=';')
to_train = pd.read_csv('for_algorithm.csv', sep=';')
X = to_train.iloc[:, :25].values
y_regression = to_train.iloc[:, 26].values
y_classification = to_train.iloc[:, 25].values
linear, logistic = LinearRegression(), LogisticRegression(random_state=0)
linear.fit(X, y_regression)
logistic.fit(X, y_classification)
test_regression = np.identity(25)
test_classification = pd.DataFrame(np.random.randint(0, 26, size=(20, 3)))
for c in test_classification.columns:
    test_classification[c] = test_classification[c].map(randomizer.mapping)
test_classification_ = pd.DataFrame(data=np.zeros(shape=(len(test_classification), len(randomizer.sparse_columns))), 
                                    columns=randomizer.sparse_columns)
for i in range(len(test_classification)):
    for pizza in list(test_classification.loc[i, :]):
        if pizza in randomizer.sparse_columns:
            test_classification_.loc[i, pizza] += 1
               
test_classification_ = test_classification_.iloc[:, :].values
```

```python
y_predicted_classification = logistic.predict(test_classification_)
y_predicted_classification = pd.Series(y_predicted_classification).map(randomizer.country_map)
y_predicted_classification_with_probabilities = logistic.predict_proba(test_classification_)
outcome = pd.DataFrame(test_classification)
outcome['pred country'] = y_predicted_classification
outcome_by_countries = pd.DataFrame(y_predicted_classification_with_probabilities*100, 
                                   columns=['Spain', 'Germany', 'Great Britain', 'Finland', 'Italy'])
```
