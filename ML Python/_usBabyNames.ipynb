```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
#plotly
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
import cufflinks
cufflinks.go_offline(connected=True)
init_notebook_mode(connected=True)
```

```python
path = './babynames/'
all_files = os.listdir(path)
```

```python
all_files
```

```python
pieces = []
columns = ['name', 'sex', 'births']
for f in all_files:
    if f.endswith('.txt'):
        frame = pd.read_csv(f'{path}{f}', names=columns)
        frame['year'] = f[3:7]
        pieces.append(frame)
```

```python
assert len(pieces) == len(all_files) - 1
```

```python
names = pd.concat(pieces, ignore_index=True)
```

```python
names.shape
```

```python
total_births = names.pivot_table(values='births', index='year', columns='sex', aggfunc=sum)
total_births.tail()
```

```python
# Standard plotly imports
```

```python
total_births.iplot(mode='lines')
```

```python
def add_group(group):
    group['prop'] = group.births / group.births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_group)
```

```python
names
```

```python
names.groupby(['year','sex']).prop.sum()
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]
```

```python
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
```

```python
top1000.reset_index(inplace=True, drop=True)
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop', ascending=False)
    return group.prop.cumsum().values.searchsorted(q) + 1
```

```python
diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
```

```python
diversity.head()
```

```python
diversity = diversity.unstack('sex')
```

```python
diversity[:10]
```

```python
diversity.iplot(mode='lines', title='Diversity')
```
